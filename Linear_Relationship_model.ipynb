### **Basic example of Machine Learning using Keras**

- Problem Statement- Predict y value for x value (input value)

*The neural network architechture*

```python
from tensorflow import keras
from tensorflow.keras import layers
model=keras.Sequential([layers.Dense(units=1,input_shape=[1])])
```

*The compilation step*

```python
model.compile(optimizer='sgd',loss='mean_squared_error')
```

```python
import numpy as np
x=np.array([-3.0,-2.0,-1.0,0.0,1.0,2.0,3.0],dtype=float)
y=np.array([ -17.0,-12.0,-7.0,-2.0,3.0,8.0,12.0],dtype=float)
```

*Fitting the model*

```python
model.fit(x,y,epochs=500)
```

```python
print(model.predict([8.0]))
```

Observation:-

The answers we fed to train our model had the rule  y=5x-2 and based on the rule, the answer should have been 38. Here, we predict our result to be 37.000027 which is close to 38 but not the exact value, since we fed only 7 numbers to our network which assumes a very high probability of having a linear relationship between x's and y's but we are not certain to have a linear relationship outside of the range provided to the network.
